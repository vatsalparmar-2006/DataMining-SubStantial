### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
original_data = df.copy()
original_data['Age'] = original_data['Age'].interpolate()
original_data

# using min-max to normalization
data_minmax = original_data.copy()
age_mm = data_minmax['Age']
age_min = age_mm.min()
age_max = age_mm.max()

df['NewAge'] = (df['Age'] - age_min) / (age_max - age_min)
df
```

```python
# using decimal to normalizaion
data_decimal = original_data.copy()
age_decimal = data_decimal['Age']
max_abs_age = abs(age_decimal).max()
j = len(str(int(max_abs_age)))      # count digits in max absolute value

data_decimal['DecimalAge'] = age_decimal / (10 ** j)
data_decimal
```

```python
# using z score to normalize
data_z_score = original_data.copy()
age_z_score = data_z_score['Age']
age_mean = age_z_score.mean()
age_std = age_z_score.std()

data_z_score['ZScoreAge'] = (age_z_score - age_min) / age_std
data_z_score
```
