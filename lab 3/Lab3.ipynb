### 2)  Identify Nominal, Ordinal, Binary and Numeric attributes from data sets and display all values.



```python
# only string value (meaning less)
nominal = ['Name', 'Ticket', 'Cabin', 'Embarked'] 

# categorical value (meaning more)
ordinary = ['Pclass']

# binary values (boolean answers)
binary = ['Survived', 'Sex' ]

# numerical values (meaning most)
numeric = ['PassengerId', 'Age', 'SibSp', 'Parch', 'Fare']

print(f"Nominal : {nominal}")
print(f"Ordinary : {ordinary}")
print(f"Binary : {binary}")
print(f"Numeric : {numeric}")
```

### 3)  Identify symmetric and asymmetric binary attributes from data sets and display all values.



```python
symmetric = ['Sex']
asymmetric = ['Survived']

print(f"symmetric : {symmetric}")
print(f"asymmetric : {asymmetric}")
```

### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
numeric = ['PassengerId', 'Age', 'SibSp', 'Parch', 'Fare']

for i in numeric:
    print(f"{i} : ")
    print(f"    Mean = {df[i].mean()}")
    print(f"    Standard deviation = {df[i].std()}")
    print(f"    Minimum = {df[i].min()}")
    print(f"    Maximum = {df[i].max()}")
    print(f"    Mode = {df[i].mode()[0]}")
    print(f"    Range = {(df[i].max()) - (df[i].min())}")

# Mode -> return list of outputs 
```

### 6)  For the qualitative attribute (class), count the frequency for each of its distinct values.



```python
qualitative = ['Name', 'Sex', 'Ticket', 'Pclass', 'Embarked', 'Cabin']
for i in qualitative:
    print('----------')
    print(df[i].value_counts())
```

### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,3))
plt.hist(df['Age'], bins=8)
plt.grid()
plt.show()
```

### 10) A boxplot can also be used to show the distribution of values for each attribute.

```python
plt.figure(figsize=(4, 3))
df[['Age', 'Fare']].boxplot()
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
plt.figure(figsize=(10, 8))

# Pair 1: Age vs Fare
plt.subplot(2, 3, 1)
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')

# Pair 2: Age vs Pclass
plt.subplot(2, 3, 2)
plt.scatter(df['Age'], df['Pclass'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Pclass')
plt.title('Age vs Pclass')

# Pair 3: Fare vs Pclass
plt.subplot(2, 3, 3)
plt.scatter(df['Fare'], df['Pclass'], alpha=0.5)
plt.xlabel('Fare')
plt.ylabel('Pclass')
plt.title('Fare vs Pclass')

# Pair 4: Age vs SibSp
plt.subplot(2, 3, 4)
plt.scatter(df['Age'], df['SibSp'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('SibSp')
plt.title('Age vs SibSp')

# Pair 5: Fare vs Parch
plt.subplot(2, 3, 5)
plt.scatter(df['Fare'], df['Parch'], alpha=0.5)
plt.xlabel('Fare')
plt.ylabel('Parch')
plt.title('Fare vs Parch')

plt.tight_layout()
plt.show()
```
